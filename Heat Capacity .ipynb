```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy.optimize import minimize
from scipy.optimize import curve_fit 
#these are the packages I need for exceuting the equation and the plot function.
```

```python
T=np.arange(100, 1000.0001, 100)
print(T)
#temperature from 100 degree celsius to 1000 degree celsius by 100 increments.
CO2=np.array([36.11,0.04233,-0.00002887,7.464E-09])
H2O=np.array([33.46,0.00688,0.000007604,-3.593E-09])
O2=np.array([29.1,0.01158,-0.000006076,1.311E-09])
N2=np.array([29,0.002199,0.00000573,-2.871E-09])
```

```python
HeatCapacity1=CO2[0]+CO2[1]*T+CO2[2]*np.power(T,2)+CO2[3]*np.power(T,3)
HeatCapacity2=H2O[0]+H2O[1]*T+H2O[2]*np.power(T,2)+H2O[3]*np.power(T,3)
HeatCapacity3=O2[0]+O2[1]*T+O2[2]*np.power(T,2)+O2[3]*np.power(T,3)
HeatCapacity4=N2[0]+N2[1]*T+N2[2]*np.power(T,2)+N2[3]*np.power(T,3)
#equation for Heat Capacity for each compound at a given temperature
print(HeatCapacity1)
print(HeatCapacity2)
print(HeatCapacity3)
print(HeatCapacity4)
#print all table to compare to excel calculations




```

```python
plt.show()
plt.plot(T,HeatCapacity1,label = 'CO2')
plt.plot(T,HeatCapacity2,label= 'H2O')
plt.plot(T,HeatCapacity3,label= 'O2')
plt.plot(T,HeatCapacity4,label= 'N2')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Heat Capacity (J*Mol/C)')
plt.title('Heat Capacity Over Temperature')
plt.legend()
plt.savefig("Problem1HeatCapacity.png")
```
